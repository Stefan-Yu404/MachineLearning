And the numeric variables include `customerId`, `creditLimit`, `availableMoney`, `transactionAmount`, `posEntryMode`, `posConditionCode`, `cardCVV`, `enteredCVV`, `cardLast4Digits` and `currentBalance`

Then I draw each of those's historgram plot try to identify outliers.

- customerId: This is just a unique identifier for the customer associated with the transaction and it seems normal

- creditLimit : This variable represents the maximum credit limit that the customer has on their account. While the value of 50000 appears to be an outlier based on the plot, it is possible that in real life, the next level of credit limit beyond 20000 would likely be set at 50000.

- availableMoney: This variable represents the amount of credit available to the customer at the time of the transaction. From the plot there might exist some outliers, but considering the creditLimit, I think I only need to remove the value that is less than 0.

- posEntryMode: the method used by the customer to enter their payment card information during the transaction. There are some outliers also with value too different then the origin value.

- posConditionCode: the condition of the point-of-sale terminal at the time of the transaction. There are some outliers also.

- currentBalance: 

After all the analysis, I plan to remove all the outliers.

```python
df_now = df_now[df_now['availableMoney'] >= 0]
df_now = df_now[df_now['posEntryMode'] <= 60]
df_now = df_now[df_now['posConditionCode'] <= 60]
df_now = df_now[df_now['currentBalance'] <= 20000]
```

```python
import datetime
df_now['day_of_week'] = df_now['transactionDateTime'].apply(lambda x : 
                        datetime.datetime.strptime(x, '%Y-%m-%dT%H:%M:%S').strftime("%A"))
df_now["transactionDateTime"] = df_now["transactionDateTime"].astype("datetime64")
df_now["accountOpenDate"] = df_now["accountOpenDate"].astype("datetime64")
df_now["dateOfLastAddressChange"] = df_now["dateOfLastAddressChange"].astype("datetime64")
df_now["currentExpDate"] = df_now["currentExpDate"].astype("datetime64")
df_now["cardLast4Digits"] = [("{:>4}".format(s)).replace(" ", "0") for s in df_now.cardLast4Digits]
df_now["cardCVV"] = [("{:>3}".format(s)).replace(" ", "0") for s in df_now.cardCVV]
df_now["enteredCVV"] = [("{:>3}".format(s)).replace(" ", "0") for s in df_now.enteredCVV]
df_now["CVVMatched"] = (df_now.cardCVV == df_now.enteredCVV).replace("True", 1).replace("False", 0)
```

```python
df_now['Actual_length_cardLast4Digits'] = df_now['cardLast4Digits'].apply(lambda x : len(str(x)))
sum(df_now['Actual_length_cardLast4Digits']==4)
```

```python
df_now.drop(df_now[df_now['cardCVV'] != df_now['enteredCVV']].index, inplace=True)
df_now.drop(df_now[df_now['Actual_length_cardLast4Digits']==3].index, inplace=True)
df_now
```

```python
df_now['Matched_CVV'] = df_now.apply(lambda row: row['cardCVV'] == row['enteredCVV'], axis=1)
grouped1 = df_now.groupby([ 'Matched_CVV'])['isFraud'].mean().reset_index()
import seaborn as sns
sns.catplot(x='Matched_CVV', y='isFraud', kind='bar', data=grouped1)
```

```python
df_now['NOT_Matched_CVV'] = df_now.apply(lambda row: row['cardCVV'] != row['enteredCVV'], axis=1)
grouped2 = df_now.groupby([ 'NOT_Matched_CVV'])['isFraud'].mean().reset_index()
import seaborn as sns
sns.catplot(x='NOT_Matched_CVV', y='isFraud', kind='bar', data=grouped2)
```

```python
g = sns.catplot(x='Matched_CVV', y='isFraud', kind='bar', data=grouped1)
g.set_axis_labels('Correct CVV', 'Fraud rate')
g = sns.catplot(x='NOT_Matched_CVV', y='isFraud', kind='bar', data=grouped2)
g.set_axis_labels('InCorrect CVV', 'Fraud rate')

# Combine plots
plt.subplots_adjust(wspace=0.4)
plt.show()
```

```python
list = ["creditLimit","merchantCategoryCode", "posEntryMode", "transactionType", "posConditionCode", "merchantCountryCode"]
list2 = ["posEntryMode", "posConditionCode"]
sns.set(style="whitegrid")
for var in list2:
    fraud_rate_by_mcc = df_now.groupby(var)['isFraud'].mean().sort_values()
    plt.bar(fraud_rate_by_mcc.index, fraud_rate_by_mcc.values, width=4)
    plt.xticks(rotation=60)
    plt.title(f'Fraud Rate by {var}')
    plt.xlabel(f'{var}')
    plt.ylabel('Fraud Rate')
    plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 8))
axs = axs.flatten()
list2 = ["posEntryMode", "posConditionCode"]
for i, var in enumerate(list2):
    fraud_rate_by_mcc = df_now.groupby(var)['isFraud'].mean().sort_values()
    axs[i].bar(fraud_rate_by_mcc.index, fraud_rate_by_mcc.values, width=4)
    axs[i].set_xticklabels(fraud_rate_by_mcc.index, rotation=70)
    axs[i].set_title(f'Fraud Rate by {var}')
    axs[i].set_xlabel(f'{var}')
    axs[i].set_ylabel('Fraud Rate')
plt.tight_layout()
plt.show()
```

Analyzing the pattern of the visualization, we can gain insights into which categories have a higher or lower likelihood of being associated with fraudulent transactions. For example, we may observe that certain merchant categories have a higher fraud rate compared to others, or that certain types of transactions have a higher fraud rate. These insights can be useful in creating a predictive model for fraudulent transactions, as they can inform the selection of features to include in the model and help identify potential areas of focus for fraud detection and prevention efforts.

```python
grouped = df_now.groupby(['merchantCategoryCode', 'transactionType'])['isFraud'].mean().reset_index()
ax = sns.catplot(x="merchantCategoryCode", y="isFraud", hue="transactionType", data=grouped, kind="bar", height=6, aspect=2)
ax.set(xlabel='Merchant Category Code', ylabel='Fraud Rate', title='Fraud Rate by Merchant Category Code and Transaction Type')
plt.xticks(rotation=60)
plt.show()
```

```python
grouped = df_now.groupby(['merchantCategoryCode', 'transactionType'])['isFraud'].mean().reset_index()
ax = sns.catplot(x="merchantCategoryCode", y="isFraud", hue="transactionType", data=grouped, kind="bar", height=6, aspect=2, palette=['#0072B2', '#E69F00', '#009E73'])
ax.set(xlabel='Merchant Category Code', ylabel='Fraud Rate', title='Fraud Rate by Merchant Category Code and Transaction Type')
plt.xticks(rotation=60)
for p in ax.ax.patches:
    ax.ax.annotate(format(p.get_height(), '.2%'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.show()
```

The resulting chart can provide insights into which combinations of merchantCategoryCode and transactionType are associated with higher fraud rates. For example, we may observe that some merchantCategoryCode values have consistently higher fraud rates across different transactionType values, suggesting that they may be more vulnerable to fraud. Alternatively, we may observe that certain transactionType values are associated with higher fraud rates across different merchantCategoryCode values, indicating that they may be riskier transaction types overall. These insights can inform the development of predictive models for fraudulent transactions by highlighting which features may be most predictive of fraud.
